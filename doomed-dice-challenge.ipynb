```python
# Doomed Dice Challenge in Python
import itertools
from collections import defaultdict

# Part A - Total Combinations and Distribution
Die_A = [1, 2, 3, 4, 5, 6]
Die_B = [1, 2, 3, 4, 5, 6]

combinations = list(itertools.product(Die_A, Die_B))

distribution = defaultdict(int)
for a, b in combinations:
    distribution[a + b] += 1

total_combinations = len(combinations)

# Probability Calculation
probabilities = {key: distribution[key] / total_combinations for key in distribution.keys()}

# Part B - Loki's Challenge
def undoom_dice(die_A, die_B):
    new_die_A = [1, 1, 2, 2, 3, 4]  # Restricted to 4 spots max
    new_die_B = [1, 2, 3, 4, 5, 6]

    new_combinations = list(itertools.product(new_die_A, new_die_B))
    new_distribution = defaultdict(int)

    for a, b in new_combinations:
        new_distribution[a + b] += 1

    new_probabilities = {key: new_distribution[key] / len(new_combinations) for key in new_distribution.keys()}
    return new_die_A, new_die_B, new_probabilities

new_die_A, new_die_B, new_probabilities = undoom_dice(Die_A, Die_B)

print("New Die A:", new_die_A)
print("New Die B:", new_die_B)
print("New Probabilities:", new_probabilities)
```
